# EV07 - Reconocimiento de tipos de variables y tipos de gráficas para visualización de datos.


## 1) Conceptos base

### 1.1 Tipos de variables
- Cualitativa nominal: categorias sin orden (ejemplo: genero, plataforma).
- Cualitativa ordinal: categorias con orden (ejemplo: nivel bajo/medio/alto).
- Cuantitativa discreta: valores enteros contables (ejemplo: cantidad).
- Cuantitativa continua: valores numericos en intervalos (ejemplo: ventas, peso, tiempo).

### 1.2 Tipos de graficas (uso recomendado)
- Barras: comparar categorias.
- Pastel: composicion porcentual con pocas categorias.
- Lineas: tendencias en el tiempo.
- Histograma: distribucion de variable numerica continua.
- Dispersion: relacion entre dos variables numericas.
- Boxplot: dispersion, mediana y posibles atipicos.

## 6) Reflexion final

1. **Que variables fueron mas dificiles de clasificar y por que?**
La variable mas dificil fue `Rank`, porque aunque usa numeros, su interpretacion principal es de orden (posicion) y no de magnitud medible como una venta. Por eso se clasifica mejor como ordinal.

2. **Que riesgos genera elegir una grafica incorrecta?**
Una grafica inadecuada puede ocultar patrones, exagerar diferencias o llevar a conclusiones equivocadas. Por ejemplo, usar pastel con muchas categorias dificulta comparar valores y puede inducir decisiones mal fundamentadas.

3. **Como impacta la clasificacion de variables en la analitica de datos?**
La clasificacion correcta define que tecnicas y visualizaciones son validas. Si una variable se clasifica mal, se aplican metodos incorrectos y disminuye la calidad del analisis, afectando la interpretacion y la toma de decisiones.
