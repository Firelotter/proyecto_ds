## 5. Selección y filtrado

```python
if "pd" not in globals():
    import pandas as pd
if "display" not in globals():
    from IPython.display import display
if "df" not in globals():
    ruta_dataset = "../data/raw/dataset.csv"
    df = pd.read_csv(ruta_dataset)
    print(f"Se recargo df desde: {ruta_dataset}")

# Seleccion de columnas especificas
columnas_clave = ["Name", "Platform", "Year", "Genre", "Global_Sales"]
print("Vista de columnas clave:")
display(df[columnas_clave].head(8))

# Seleccion por etiquetas con loc (filas 0 a 5 y columnas especificas)
print("Seleccion con loc:")
display(df.loc[0:5, ["Name", "Platform", "Year", "Global_Sales"]])

# Seleccion por posicion con iloc (primeras 5 filas y primeras 4 columnas)
print("Seleccion con iloc:")
display(df.iloc[0:5, 0:4])

# Filtro 1: videojuegos lanzados desde 2010
filtro_2010 = df[df["Year"] >= 2010]
print(f"Registros con Year >= 2010: {filtro_2010.shape[0]}")
display(filtro_2010[["Name", "Year", "Global_Sales"]].head(10))

# Filtro 2: genero Action con ventas globales mayores a 1.0
filtro_action = df[(df["Genre"] == "Action") & (df["Global_Sales"] > 1.0)]
print(f"Action con Global_Sales > 1.0: {filtro_action.shape[0]}")
display(filtro_action[["Name", "Platform", "Genre", "Global_Sales"]].head(10))

# Filtro 3: registros con Publisher faltante
publisher_nulos = df[df["Publisher"].isna()]
print(f"Registros con Publisher nulo: {publisher_nulos.shape[0]}")
display(publisher_nulos[["Name", "Platform", "Year", "Publisher"]].head(10))
```

## 6. Manipulación básica

```python
if "pd" not in globals():
    import pandas as pd
if "display" not in globals():
    from IPython.display import display
if "df" not in globals():
    ruta_dataset = "../data/raw/dataset.csv"
    df = pd.read_csv(ruta_dataset)
    print(f"Se recargo df desde: {ruta_dataset}")

# Parte A: ejemplos de creacion de DataFrame (alineado con la presentacion)
df_dict = pd.DataFrame({
    "Nombre": ["Ana", "Luis", "Marta"],
    "Edad": [23, 35, 29],
    "Ciudad": ["Madrid", "Barcelona", "Valencia"]
})
print("DataFrame creado desde diccionario:")
display(df_dict)

df_lista = pd.DataFrame(
    [["Carlos", 31, "Bogota"], ["Diana", 27, "Medellin"]],
    columns=["Nombre", "Edad", "Ciudad"]
)
print("DataFrame creado desde lista de listas:")
display(df_lista)

# Parte B: manipulacion basica sobre el dataset real
df_trabajo = df.copy()

# Crear columna calculada con suma de ventas regionales
df_trabajo["Total_Regional_Sales"] = (
    df_trabajo["NA_Sales"]
    + df_trabajo["EU_Sales"]
    + df_trabajo["JP_Sales"]
    + df_trabajo["Other_Sales"]
)

# Crear una columna de control para comparar contra Global_Sales
df_trabajo["Diferencia_vs_Global"] = (df_trabajo["Global_Sales"] - df_trabajo["Total_Regional_Sales"]).round(4)

# Renombrar columnas para claridad
df_trabajo = df_trabajo.rename(columns={"Name": "Game_Name", "Year": "Release_Year"})

# Completar faltantes en Publisher con etiqueta controlada
df_trabajo["Publisher"] = df_trabajo["Publisher"].fillna("Unknown")

# Eliminar columna temporal de control
df_trabajo = df_trabajo.drop(columns=["Diferencia_vs_Global"])

# Ordenar por ventas globales de mayor a menor
df_ordenado = df_trabajo.sort_values(by="Global_Sales", ascending=False)

print("Top 10 juegos por ventas globales (despues de manipular):")
display(df_ordenado[["Game_Name", "Platform", "Release_Year", "Genre", "Global_Sales", "Total_Regional_Sales"]].head(10))

# Resumen por genero para mostrar una transformacion adicional
resumen_genero = (
    df_ordenado.groupby("Genre", as_index=False)["Global_Sales"]
    .mean()
    .sort_values(by="Global_Sales", ascending=False)
)
print("Promedio de Global_Sales por genero:")
display(resumen_genero)
```
